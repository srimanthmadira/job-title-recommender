```python
import pandas as pd
```

```python
job_titles = pd.read_csv("unique_job_titles.csv")
```

```python
import re

def clean_job_title(title):
    title = re.sub(r'\(.*?\)|\[.*?\]', '', title)

    title = re.sub(r'[\d,]', '', title)

    title = re.sub(r'\s+', ' ', title).strip()

    return title

job_titles['cleaned_job_titles'] = job_titles['Unique Job Titles'].apply(clean_job_title)
```

```python
import pandas as pd
import re

unwanted_terms = {"fulltime", "parttime", "contract", "temporary", "remote", "onsite", "permanent"}

def clean_job_title_simple(title):
    title = re.sub(r'\s*-\s*.*$', '', title)

    title_words = title.split()
    cleaned_words = [word for word in title_words if word.lower() not in unwanted_terms]

    cleaned_title = " ".join(cleaned_words).strip()
    return cleaned_title

job_titles['further_cleaned_job_titles'] = job_titles['cleaned_job_titles'].apply(clean_job_title_simple)
```

```python
job_titles.head()
```

```python
import pandas as pd

# Load the CSV files into DataFrames
dice_df = pd.read_csv('dice_com-job_us_sample.csv')  # This should contain the jobtitle column

# Create a mapping dictionary from unique_job_titles DataFrame
mapping = dict(zip(job_titles['Unique Job Titles'], job_titles['further_cleaned_job_titles']))

# Replace the jobtitle column in the dice_df DataFrame with the cleaned titles
dice_df['jobtitle'] = dice_df['jobtitle'].replace(mapping)

# Display the updated DataFrame
print("\nUpdated Dice DataFrame:")
print(dice_df.head())

# Optionally save the updated DataFrame back to CSV
dice_df.to_csv('dice_job_data.csv', index=False)
```
